### Adicionando marcadores

```python
mapa = folium.Map(
    location=[-23.3919536,-46.453279],
    tiles='Stamen Terrain',
    zoom_start=14
)
```

```python
# Adicionando um icone sobre o mapa localizando o Aeroporto
folium.Marker(
    [-23.4378476,-46.4872113],
    popup='<i>Aeroporto Internacional de GRU</i>',
    tooltip='Clique Aqui!'
).add_to(mapa)
```

```python
mapa
```

```python
# Adicionará um icone sobre o mapa localizando o Aeroporto na cor Verde
folium.Marker(
    [-23.4378476,-46.4872113],
    popup='<i>Aeroporto Internacional de GRU</i>',
    tooltip='Clique Aqui!',
    icon=folium.Icon(color='green')
).add_to(mapa)
```

```python
mapa
```

```python
# Adicionando circulo ao redor do aeroporto
folium.CircleMarker(
    location=[-23.4378476,-46.4872113],
    radius=50,
    color='#3186cc',
    fill=True,
    fill_color='#3186cc'
).add_to(mapa)
```

```python
mapa
```

```python
import pandas as pd
from folium import plugins
```

```python
df = pd.read_csv('olist_geolocation_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
coordenadas=[]
for lat, lng in zip(df.geolocation_lat.values[:18000],df.geolocation_lng.values[:18000]):
    coordenadas.append([lat,lng])
```

```python
# Exibindo um mapa mais amplo
mapa = folium.Map(location=[-15.788497,-47.879873],zoom_start=5,tiles='Stamen Toner')
```

```python
mapa.add_child(plugins.HeatMap(coordenadas))
mapa
```

### Visualizando os dados mais distribuidos no mapa

```python
# Verificando os estados Unicos
df.geolocation_state.unique()
```

```python
# Contando a quantidade de estados existentes
df.geolocation_state.value_counts()
```

```python
# Criando um segundo dataframe para pegar valores de forma aleatória
# Pego uma fração de 0.03 (3%)
df2 = df.sample(frac = 0.03)
```

```python
coordenadas=[]
for lat, lng in zip(df2.geolocation_lat.values,df2.geolocation_lng.values):
    coordenadas.append([lat,lng])
```

```python
mapa = folium.Map(location=[-15.788497,-47.879873],zoom_start=5,tiles='Stamen Toner')
```

```python
mapa.add_child(plugins.HeatMap(coordenadas))
mapa
```

### Persistindo o mapa em disco

```python
mapa.save('mapa-consumidor.html')
```
